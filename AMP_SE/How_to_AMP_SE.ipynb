### $\color{Darkred}{\text{Analysis of the MSE}}$ 

We analyze the Mean Squared Error achieved by AMP and its associated State Evolution for the Subspace clustering problem. 

##### $\color{Purple}{\text{State Evolution}}$ 

We initialize SE equations in two ways: informed and uninformed; this is reflected into different choices for the parameters *$\color{darkblue}{\text{init_u,init_v}}$* corresponding to the initial overlaps $(m_u,m_v)$. The Python function implementing the iteration of the SE equation is *$\color{darkblue}{\text{iterate_UV}}$*. We fix the sparsity level *$\color{darkblue}{\rho=0.5}$* and the aspect ration *$\color{darkblue}{\alpha = 2}$*. We consider different value of the SNR and compute for each the MSE. 

```python
coefs = np.linspace(0.1,2.5,500) ; alpha = 2 ; damping = 0 ; tol = 10e-12 ; rho = 0.5 
se_inf , se_uninf = [] , []
for coef in coefs:
    # Rescale the snr according to the sparsity level
    snr = coef/(rho*np.sqrt(alpha))
    # Informed
    init_u , init_v = 1,rho   
    iteration = iterate_UV(init_u,init_v, tol=tol, max_steps = 1000,
                snr=snr,  rho=rho,alpha=alpha,damping=damping )
    # Uninformed
    init_u_un , init_v_un = 10e-4,10e-4 
    iteration_un = iterate_UV(init_u_un,init_v_un, tol=tol, max_steps = 1000,
                                                snr=snr, rho=rho,alpha=alpha,damping=damping)
    se_inf.append(iteration[2]) ; se_uninf.append(iteration_un[2]) 
```

##### $\color{Purple}{\text{AMP}}$ 

We analyze the performance of AMP at fixed *$\color{darkblue}{(n,d)=(2000,1000)}$*. We generate the data matrix using the function *$\color{darkblue}{\text{get_instance}}$* using the same parameter as in the SE iteration above *$\color{darkblue}{(\alpha,\rho)=(2,0.5)}$*. We average over 10 *$\color{darkblue}{\text{seeds}}$* and use *$\color{darkblue}{\text{damping}}$* to stabilize the iteration. We store the converged runs and compute the MSE and the statistical error, as measured by the standard deviation.

```python
dimension = 1000 ; nsamples = 2000  ; rho = 0.5 ; damping = 0.7 ; alpha = 2 ; seeds = 10
coefs_amp = np.linspace(0.1,2.5,10)
amps = [] ; errs = [] 
for coef in coefs_amp:
    # Rescale the snr according to the sparsity level
    snr = coef/(rho*np.sqrt(alpha)) ; aux = []
    for seed in range(seeds):
        truth = get_instance(rho=rho, dimension=dimension, nsamples=nsamples, snr=snr)
        init = 1e-3*np.random.randn(nsamples) , 1e-3*np.random.randn(dimension)
        run = lowRAMP(truth,init, tol=1e-7, max_steps=500, snr=snr, rho=rho,damping=damping)
        if run[1]:
            # Store converged runs 
            aux.append(run[0])
    # Take mean and standard deviation over many iterations to access AMP performance
    errs.append(np.std(aux)/(np.sqrt(seeds))) ; amps.append(np.mean(aux))
```

##### $\color{Purple}{\text{Plots}}$ 

We plot the performance of AMP and SE as a function of the SNR. The SE equations are valid in the infinite size limit $n,d \to \infty$. Despite this, we observe from the plot that the AMP performance, on a factor graph with $(n,d) = (2000,1000)$, agrees with its SE equations. There is no difference, for the values of the sparsity level chosen, between informed and uninformed initialization.

```python
plt.figure(figsize=(10,7))
plt.title(fr'MSE at fixed sparsity $\rho$ = {rho}',fontsize = 20)
plt.errorbar(x=coefs_amp,y=np.array(amps),yerr=errs,fmt='x',label='AMP')
plt.plot(coefs,np.array(se_inf),label='SE Informative')
plt.plot(coefs,np.array(se_inf),label='SE Uninformative')
plt.legend(fontsize = 20.5,frameon = False , ncol = 1 , loc = 'lower left')
plt.rc('xtick', labelsize=20)  
plt.rc('ytick', labelsize=20)
plt.rc('axes', labelsize=20) 
plt.xlabel(r"$\lambda \sqrt{\alpha}  $") ; plt.ylabel('MSE')
```

### $\color{Darkred}{\text{Thresholds}}$ 

In this section we compute the *thresholds* separating different *reconstruction phases*. We exploit the definitions we give in the paper. The spinodal transitions $\{\lambda_{alg Bayes},\lambda_{dyn}\}$ are computed as critical points of the curve describing the Signal-to-Noise ratio needed to achieve a given overlap $m_u$, i.e. $\lambda(m_u)$. The IT transition $\lambda_{it}$ is found numerically by finding when the informative fixed point of the SE equations has the same free energy as the uninformative one. The algorithmical transition is always fixed at $\lambda_{alg}=1$ and defines the stability of the trivial fixed point.

```python
rhos = [0.09,0.11,0.13,0.15] ; tol = 1e-4
alpha = 2
ms = np.linspace(10e-4,0.2,100) ; lamb_curves = [] ; lambs_it_new = [] ; lambs_dyn_new, lambs_alg_new = [],[]
for rho in rhos:
    lambs = [] ; deltaphis = []
    for val in ms:
        # Find lambda(mu)
        snr_sc = find_sol(val,alpha,rho,tol)
        lambs.append(snr_sc[0]*(np.sqrt(alpha)*rho))
        # Find free energy between trivial fixed point and magnetization mu
        var = deltaphi(val,snr_sc,alpha,rho)
        deltaphis.append(var)
        
    # Store lambda(mu) if not many rhos   
    lamb_curves.append(lambs)

    # Check when free energy changes sign to detect IT transition
    idx = np.where(np.sign(deltaphis[:-1]) != np.sign(deltaphis[1:]))[0] + 1
    it_trans = lambs[idx[len(idx)-1]]
    lambs_it_new.append(it_trans)
    if rho<0.07:
        alg_tr = 1
    else:
        peaks_alg, _ = find_peaks(np.array(lambs)) ; alg_tr = lambs[peaks_alg[0]] 
    peaks_dyn, _ = find_peaks(-np.array(lambs)) ; dyn_tr = lambs[peaks_dyn[0]] ; lambs_dyn_new.append(dyn_tr)
    lambs_alg_new.append(alg_tr)
    print(f'rho={rho} lamb_it = {it_trans} lamb_alg = {alg_tr} dyn_tr = {dyn_tr}')
```

```python
chars = ['d','c','b','a']
fig ,ax = plt.subplots(2,2,figsize=(16,13))

for i in range(len(rhos)):
    k = int(i<2) ; j = 1 - i % 2
    ax[k,j].plot(ms,lamb_curves[i])    
    ax[k,j].axhline(y=lambs_alg_new[i],linestyle = '-.',c='yellow',label=r'$\lambda_{alg-Bayes}$')
    ax[k,j].axhline(y=1,linestyle = '-',c='m',label=r'$\lambda_{alg}$')
    ax[k,j].axhline(y=lambs_it_new[i],linestyle = '--',c='c',label=r'$\lambda_{it}$')
    ax[k,j].axhline(y=lambs_dyn_new[i],linestyle = 'dotted',c='black',label=r'$\lambda_{dyn}$')
    ax[k,j].set_title(fr' $\rho =  $ {rhos[i]}',fontsize=18,loc='left',fontstyle = 'italic')
    ax[k,j].set_ylabel(r'$\lambda \sqrt{\alpha}\, $') ; ax[k,j].set_xlabel(r'$m_u$') 
    ax[k,j].set_ylim(0.8,1.05) ; ax[k,j].set_xlim(-1e-10,0.2)
    if k == 1:
        ax[k,j].legend(fontsize = 18, loc = 'upper right')
    else:
        ax[k,j].legend(fontsize = 18, loc = 'lower left') 
plt.rc('xtick', labelsize=15)  
plt.rc('ytick', labelsize=15)
plt.rc('axes', labelsize=18) 
```
